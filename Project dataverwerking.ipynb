```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# De nagalmtijd per punt in een lijst zetten.
rij1= [1.05,1.42,1.02,1.45,1.20,1.29,0.91]
rij2= [0.98,0.75,1.05,1.09,1.28,0.51,0.81]
rij3= [1.21,1.36,1.04,1.25,0.82,1.11,1.38]
rij4= [1.06,1.02,1.02,1.03,1.03,0.95,1.12]
rij5= [1.71,1.75,1.30,1.28,1.47,1.21,1.06]
rij6= [1.26,1.34,1.65,1.73,1.54,1.08,1.60]
rij7= [1.63,1.32,1.12,1.29,1.31,1.28,0.94]
```

```python
# De rijen bij elkaar toevoegen.
rijtot= rij1,rij2,rij3,rij4,rij5,rij6,rij7
```

```python
x = [0,1,2,3,4,5,6,7] # De x coördinaten invoeren (zodat je een grid kan maken)
y = [0,1,2,3,4,5,6,7] # De y coördinaten invoeren (zodat je een grid kan maken)
X, Y = np.meshgrid(x,y)
Z = rijtot # De lijst die op de meshgrid gezet gaat worden.

# plot
plt.style.use('classic')
plt.figure(dpi=100)
ax = plt.axes()

plt.pcolormesh(X, Y, Z, vmin=0.01, vmax=2.5) # Met de getallen kan de intensiteit van de colorbar aanpassen.
plt.colorbar(label="nagalmtijd")
#plt.plot(X, Y, '.')
#plt.contour(x, y, Z,)

plt.show()
```
